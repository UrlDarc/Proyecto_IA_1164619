```python
import re


df['text'] = df['text'].fillna('')


def clean_text(text):
    if not isinstance(text, str):  
        return ''  
    text = text.lower()  
    text = re.sub(r'http\S+', '', text)  
    text = re.sub(r'@\w+', '', text) 
    text = re.sub(r'#\w+', '', text)  
    text = re.sub(r'[^a-zA-Z\s]', '', text) 
    return text


df['cleaned_text'] = df['text'].apply(clean_text)


print(df.head())

```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()


def preprocess_text(text):
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words and word.isalpha()]
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(lemmatized_tokens)


df['lemmatized_text'] = df['cleaned_text'].apply(preprocess_text)


print(df[['text', 'cleaned_text', 'lemmatized_text']].head())
```

```python
import random

data = list(zip(df['lemmatized_text'], df['sentiment']))

random.shuffle(data)

split_idx = int(0.8 * len(data))
train_data = data[:split_idx]
test_data = data[split_idx:]
```

```python
from collections import defaultdict
import math

class_counts = defaultdict(int)
word_counts = defaultdict(lambda: defaultdict(int))
vocab = set()

for text, label in train_data:
    class_counts[label] += 1
    words = text.split()
    for word in words:
        word_counts[label][word] += 1
        vocab.add(word)

total_docs = sum(class_counts.values())


log_probs = {}
class_priors = {}

for label in class_counts:
    total_words = sum(word_counts[label].values())
    class_priors[label] = math.log(class_counts[label] / total_docs)
    log_probs[label] = {}

    for word in vocab:
        word_freq = word_counts[label][word] + 1 
        log_probs[label][word] = math.log(word_freq / (total_words + len(vocab)))
```

```python
def predict(text):
    words = text.split()
    scores = {}

    for label in class_priors:
        scores[label] = class_priors[label]
        for word in words:
            if word in vocab:
                scores[label] += log_probs[label].get(word, math.log(1 / (sum(word_counts[label].values()) + len(vocab))))
    
    return max(scores, key=scores.get)
```

```python
correct = 0
total = len(test_data)

for text, label in test_data:
    prediction = predict(text)
    if prediction == label:
        correct += 1

accuracy = correct / total
print(f"Precisión del modelo Naive Bayes: {accuracy:.2f}")
```

```python
import random
import math
from collections import defaultdict

stopwords = set(["the", "and", "is", "to", "it", "of", "a", "in", "that", "for", "on", "with", "as", "was", "at", "by"])


def clean_text(text):
    words = text.split()
    return ' '.join([word.lower() for word in words if word not in stopwords])


df = pd.read_csv('Tweets.csv')


df['cleaned_text'] = df['text'].apply(clean_text)
df['lemmatized_text'] = df['cleaned_text']  


data = list(zip(df['lemmatized_text'], df['sentiment']))


random.shuffle(data)

split_idx = int(0.8 * len(data))
train_data = data[:split_idx]
test_data = data[split_idx:]

class_counts = defaultdict(int)
word_counts = defaultdict(lambda: defaultdict(int))
vocab = set()


for text, label in train_data:
    class_counts[label] += 1
    words = text.split()
    for word in words:
        word_counts[label][word] += 1
        vocab.add(word)


total_docs = sum(class_counts.values())


log_probs = {}
class_priors = {}

for label in class_counts:
    total_words = sum(word_counts[label].values())
    class_priors[label] = math.log(class_counts[label] / total_docs)
    log_probs[label] = {}

    for word in vocab:
        word_freq = word_counts[label][word] + 1  
        log_probs[label][word] = math.log(word_freq / (total_words + len(vocab)))

def predict(text):
    words = text.split()
    scores = {}

    for label in class_priors:
        scores[label] = class_priors[label]
        for word in words:
            if word in vocab:
                scores[label] += log_probs[label].get(word, math.log(1 / (sum(word_counts[label].values()) + len(vocab))))

        
        if class_counts[label] < total_docs * 0.2:  
            scores[label] *= 1.5 

    return max(scores, key=scores.get)


correct = 0
total = len(test_data)

for text, label in test_data:
    prediction = predict(text)
    if prediction == label:
        correct += 1

accuracy = correct / total
print(f"Precisión del modelo Naive Bayes: {accuracy:.2f}")

```
